```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
# Preprocessing Input data
data = {'A': [4  , 5  , 5  , 6  , 7  , 8, 8  , 9],
        'B': [4.5, 4.5, 5.3, 5.7, 7.4, 8, 8.5, 8.5]}
```

```python
data=pd.DataFrame.from_dict(data)
```

```python
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
X_mean = np.mean(X)
Y_mean = np.mean(Y)
num = 0
den = 0
```

```python
for i in range(len(X)):
    num += (X[i] - X_mean)*(Y[i] - Y_mean)
    den += (X[i] - X_mean)**2
```

```python
m = num / den
c = Y_mean - m*X_mean
print (m, c)
```

```python
# Making predictions
Y_pred = m*X + c
plt.scatter(X, Y) # actual
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
M = [[1, 1, 1, 1], [1, 2, 4, 2**3], [1, 3, 9, 27], [1, 3.4, 3.4**2, 3.4**3]]
Y = [1.1, 2.1, 5, 7]
```

```python
a = np.matmul(np.linalg.inv(M),Y)
```

```python
x = sp.symbols('x')
```

```python
y = a[0] + a[1]*x + a[2]*x**2 + a[3]*x**3
```

```python
x_val = np.arange(0,4,0.01)
```

```python
plt.plot(x_val, [y.subs(x,i) for i in x_val]) # subs=Substitute
plt.scatter([1, 2, 3, 3.4],[1.1, 2.1, 5, 7], c='k')
plt.title(sp.N(y,5))
plt.grid(); plt.show()
```

```python
y = 4.5
```

```python
X  = [1.27, 2.23, 2.5, 3.6]
Y = [2.3, 2.95, 3.5, 5.1]
```

```python
x = 0
for n in range(len(X)):
    num = 1
    den = 1
    for m in range(len(Y)):
        if m!=n:
            num *= y - Y[m]
            den *= Y[n] - Y[m]
    x += (num/den) * X[n]
print(x)
```

```python
plt.plot(X, Y)
plt.title("y = {:.1f}  and  x = {:.2f}".format(y, x))
plt.grid(); plt.show()
```
