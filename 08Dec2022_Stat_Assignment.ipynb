```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xn = []
yn = []
def f(y, x, h, lim):
    if x>=lim:
        return y
    y = y + h*(x**3 + y)
    yn.append(y)
    xn.append(x)
    return f(y, x+h, h, lim)
```

```python
print(xn)
print(yn)
print("y(0.12) = {:<0.2f}".format(f(1, 0, 0.02, 0.12)))
```

```python
plt.subplot()
plt.scatter(xn, yn)
plt.title("Euler's Method")
plt.xlabel("f(x)")
plt.ylabel("f(y)")
plt.grid()
plt.show()
```
