```python
def IT(f, a, b, n):
    h = (b - a)/n
    sum = 0
    for i in range (int(n)):
        t = (f(a + i*h) + f(a + (i + 1)*h) * h/2)
        sum = sum + t
    return sum
```

```python
def f(x):
    y = 2 + 2*x + x**2 + np.sin(2*np.pi*x) + np.cos(2*np.pi*x/0.5)
    return y
```

```python
print(IT(f, 0, 1.5, 1.0))
print(IT(f, 0, 1.5, 9.0))
```

```python
def IS1(f, a, b, n):
    h = (b - a)/(2*n)
    sum = 0
    for i in range (int(n)):
        t = (f(a + 2*i*h) + 4*f(a + (2*i + 1)*h) + f(a + (2*i + 2)*h))
        sum = sum + t
    return sum
```

# Assignment
    Input : f(x) = -x^2 + 4x + 5
    Output: -2

```python
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```python
def f(x):
    dif = -x**2 + 4*x + 5
    return dif
def f1(x):
    dif = -2*x + 4
    return dif
def f2(x) :
    if x!=0:
        return -2*(x/x)
    return 0
```

```python
X = []
Y = []
Z = []
C = []
for x in range(-2, 6):
    X.append(x)
    Y.append(f(x))
    Z.append(f1(x))
    C.append(f2(x))
print(X)
print(Y)
print(Z)
print(C)
mean = []
for i in range(len(X)):
    mean.append((Y[i]+Z[i]+C[i])/2)
print(mean)
```

```python
plt.subplot()
plt.plot(X, Y, label="f(x)")
plt.plot(X, Z, label="f1(x)")
plt.plot(X, C, label="f2(x)")
plt.plot(X, mean, label="Mean")
plt.legend()
plt.show()
```
