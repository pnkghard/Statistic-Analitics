```python
from sklearn import datasets
```

```python
dataset = datasets.load_wine() #load dataset
```

```python
print("Inputs : ", dataset.feature_names) #print 13 feature
```

```python
print("Outputs : ", dataset.target_names) #print label of wine
```

```python
print(dataset.data[0:3]) #print the wine data feature
```

```python
print(dataset.target) #print wine label
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs = dataset.data
```

```python
outputs = dataset.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, outputs, test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB #import gaussian naive bay library
```

```python
classifer = GaussianNB()
```

```python
classifer.fit(x_train, y_train) #train the model
```

```python
y_pred = classifer.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy : ", metrics.accuracy_score(y_test, y_pred))
```

```python
import seaborn as sns
sns.set_theme()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
```

```python
#Generate data x- axis
x= np.arange(-3, 3, 0.001)
```

```python
# multiple noramla distributions
plt.plot(x, norm.pdf(x, 0, 1), label='Mean:0, SD:1')
plt.plot(x, norm.pdf(x, 0, 1.5), label='Mean:0, SD:1.5')
plt.plot(x, norm.pdf(x, 0, 2), label='Mean:0, SD:2')
plt.xlabel("Standard Deviation")
plt.ylabel("Size of Data")
plt.title("Normal Distribution")
plt.legend()
```
